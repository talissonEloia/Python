```python
salto_a = float(input('Salto do atleta 1:'))
salto_b = float(input('Salto do atleta 2:'))
evola = float(input('evolução do atleta 1 (cm):'))
evolb = float(input('evolução do atleta 2 (cm):'))
cont1 = 0
cont2 = 0

print('João com salto incial de %.2f e a cada salto ele sobe %.2fcm'%(salto_a,evola))
print('José com salto incial de %.2f e a cada salto ele sobe %.2fcm'%(salto_b,evolb))

if salto_a > salto_b:
  while salto_b < salto_a:
    salto_b = salto_b + evola
    cont1 += 1 
elif salto_b > salto_a:
  while salto_b > salto_a:
    salto_a = salto_a + evolb
    cont2 += 1
else:
  
  if evola > evolb:
    print('João supera josé no salto nº 1')
  elif evola < evolb:
    print('José supera João no salto nº 1')
  else:
    print('Saltos iguais.')

print('João supera josé no salto nº %i'%cont2)

print('José supera João no salto nº %i'%cont1)
```
